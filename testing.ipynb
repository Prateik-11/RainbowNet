```python
import torch
from StartingDataset import StartingDataset 
from networks.StartingNetwork import StartingNetwork
from starting_train import starting_train
```

```python
model = torch.nn.DataParallel(StartingNetwork())
ckpt = torch.load('/localhome/advit/rainbow_train/leaf-us-alone/checkpoints/model.pt')
model_ckpt = ckpt['model']
model.load_state_dict(model_ckpt)
model = model.to('cuda')
model.requires_grad_ = False
```

```python
x = torch.rand(1,8,128,256, requires_grad=True, device='cuda')
```

```python
lr=1e2

for i in range(10000):
    y = model(x)
    loss = torch.sum((y - torch.ones_like(y))**2)
    x.retain_grad()
    loss.backward()
    # with torch.no_grad():
    #     print(f"{i}th loss: {loss}  grad: {torch.sum(x.grad)}          ", end = '\r')
    print(loss)
    print(torch.sum(x.grad))
    x = x - lr * x.grad
```

```python
softmax = torch.nn.Softmax2d()
torch.round(torch.softmax(model(x)))
```
